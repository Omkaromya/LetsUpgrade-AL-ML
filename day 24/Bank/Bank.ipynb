# created Random forest Alog

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rf_model=RandomForestClassifier(n_estimators=1000,max_features=2,oob_score=True)
```

```python
database2.columns
```

```python
feactures=['Age', 'Experience', 'Income', 'Family', 'CCAvg', 'Education',
       'Mortgage', 'Securities Account', 'CD Account',
       'Online', 'CreditCard']
```

```python
rf_model.fit(X=database2[feactures],y=database2["Personal Loan"])
```

```python
print("OOB Accuary Score:")
print(rf_model.oob_score_)
```

```python
for feacture ,imp in zip(feactures,rf_model.feature_importances_):
    print(feacture,imp)
```
